#  Numpy Array view vs Copy

```python
i=np.array([1,2,3,4,5])
x=i.copy()
x[0]=10
print(i)
print(x)
```

```python
j=np.array([1,2,3,4,5])
y=j.view()
y[0]=5
print(j)
print(y)
```

#  Check if array owns it's Data 
                                    Copies owns the data ,and views dose not owns the data

```python
k=np.array([1,2,3,4,5,6,7])
l=k.view()
m=k.copy()
print(l.base)
print(m.base)
```
